# Visulalisation

```python
import pandas as pd
import seaborn as sns
```

```python
# Load DataFrame
import pickle
path1 = './data/df_results_l1.pickle'
path2 = './data/df_results_l2.pickle'
with open(path1, 'rb') as f1, open(path2, 'rb') as f2:
    df1 = pickle.load(f1)
    df2 = pickle.load(f2)
```

```python
# Build data from both files
df_results = (pd.concat([df1, df2])
                .dropna())
```

## Desired format :

```python
# Sanity check
(df_results.epsilon < df_results.eta).sum()
df_results[df_results.epsilon == 0].sum()
```

```python
# Rename PGD n_step == 1 to FGSM
# These are FGSM since FGSM is PGD with n_step =1
mask_fgsm = df_results.n_steps == 1
df_results.loc[mask_fgsm, 'method', ] = (df_results.loc[mask_fgsm, 'method']
                                                   .str
                                                   .replace('pgd', 'fgsm'))
```

```python
# Creating two DataFrames containg l1 and l2 data
# It may looks redundant because of the initial data loading
# But we sometimes the pickles we load are not that well sorted
mask_l2 = df_results.method.str.contains('l2')
df_results_l1 = df_results[~mask_l2]
df_results_l2 = df_results[mask_l2]
```

## $L^\infty$ attacks grid.

```python
g = sns.relplot(x="epsilon", y="accuracy",
                col="method", hue='loss_name', col_wrap=2, height=4,
                col_order=['fgsm_infinity', 'fgsm_infinity_random', 'pgd_infinity', 'pgd_infinity_random'],
                kind="line", data=df_results_l1)
```

```python
g = sns.relplot(x="n_steps", y="accuracy",
                col="method", hue='loss_name', col_wrap=2, height=4,
                col_order=['pgd_infinity', 'pgd_infinity_random'],
                kind="line", data=df_results_l1.loc[~(df_results_l1.n_steps == 1)])
```

```python
g = sns.relplot(x="eta", y="accuracy",
                col="method", col_wrap=2, height=4, hue='loss_name',
                col_order=['fgsm_infinity', 'fgsm_infinity_random', 'pgd_infinity', 'pgd_infinity_random'],
                kind="line", data=df_results_l1)
```
