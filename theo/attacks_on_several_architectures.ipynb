```python
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
from attack import pgd
import numpy as np
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (20.0,8.0)
```

```python
model_32_64_128_20e = tf.keras.models.load_model("model_32_64_128_20e")
model_32_64_128_14e = tf.keras.models.load_model("model_32_64_128_14e")
model_32_64_128_dropout_14e = tf.keras.models.load_model("model_32_64_128_dropout_14e")
model_64_128_256_20e = tf.keras.models.load_model("model_64_128_256_20e")
model_64_128_256_7e = tf.keras.models.load_model("model_64_128_256_7e")
model_64_128_256_dropout_17e = tf.keras.models.load_model("model_64_128_256_dropout_17e")
```

```python
(x_train_full, y_train_full), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
x_train, x_valid = x_train_full[:40000] / 255.0, x_train_full[40000:] / 255.0
y_train, y_valid = y_train_full[:40000], y_train_full[40000:]
x_test = x_test / 255.0
y_train = tf.keras.utils.to_categorical(y_train)
y_valid = tf.keras.utils.to_categorical(y_valid)
y_test = tf.keras.utils.to_categorical(y_test)
```

```python
loss = tf.keras.losses.CategoricalCrossentropy()
```

```python
x_pgd1, list_loss1, list_accuracy1, list_interation1 = pgd(x_test, y_test, model_32_64_128_20e, loss)
```

```python
x_pgd2, list_loss2, list_accuracy2, list_interation2 = pgd(x_test, y_test, model_32_64_128_14e, loss)
```

```python
x_pgd3, list_loss3, list_accuracy3, list_interation3 = pgd(x_test, y_test, model_32_64_128_dropout_14e, loss)
```

```python
x_pgd4, list_loss4, list_accuracy4, list_interation4 = pgd(x_test, y_test, model_64_128_256_20e, loss)
```

```python
x_pgd5, list_loss5, list_accuracy5, list_interation5 = pgd(x_test, y_test, model_64_128_256_7e, loss)
```

```python
x_pgd6, list_loss6, list_accuracy6, list_interation6 = pgd(x_test, y_test, model_64_128_256_dropout_17e, loss)
```

```python
fig, ax = plt.subplots() 
ax.plot(list_interation1, list_loss1, label='loss_32_64_128_20e', c='#1f77b4')  
ax.plot(list_interation2, list_loss2, label='loss_32_64_128_14e', c='#d62728')  
ax.plot(list_interation3, list_loss3, label='loss_32_64_128_dropout_14e', c='#2ca02c') 
ax.plot(list_interation4, list_loss4, label='loss_64_128_256_20e', c='#ff7f0e')  
ax.plot(list_interation5, list_loss5, label='loss_64_128_256_7e', c='#9467bd')  
ax.plot(list_interation6, list_loss6, label='loss_64_128_256_dropout_17e', c='#8c564b')  

ax.set_xlabel('iterations')  
ax.set_ylabel('loss')  
ax.set_title("loss comparaison")  
ax.legend() 
plt.savefig('loss_evolution.png')
```
